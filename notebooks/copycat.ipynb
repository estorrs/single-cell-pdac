```r
for (l in sample_list) {
    name = l[[1]]
    fp = l[[2]]
    out_prefix = l[[3]]
    
    print(name)
    print(fp)
    print(out_prefix)
    flush.console()
    
    exp = read.table(fp, sep = "\t", header = TRUE, row.names='Gene')
    copykat.test <- copykat(
        rawmat=exp,
        id.type="S",
        ngene.chr=5,
        win.size=25,
        KS.cut=0.1,
        sam.name=name,
        distance="euclidean",
        norm.cell.names="",
        output.seg="FALSE",
        plot.genes="FALSE",
        genome="hg20",
        n.cores=20)
    
    pred.test <- data.frame(copykat.test$prediction)
    CNA.test <- data.frame(copykat.test$CNAmat)
    write.table(pred.test, file=paste(out_prefix, "_predictions.txt", sep = ""), sep='\t', quote = FALSE)
    write.table(CNA.test, file=paste(out_prefix, "_cna.txt", sep = ""), sep='\t', quote = FALSE)
}
```

```r
exp = read.table("../data/single_cell/copycat/inputs/1555-tumor.txt", sep = "\t", header = TRUE, row.names='Gene')
head(exp)
```

```r
copykat.test <- copykat(
    rawmat=exp,
    id.type="S",
    ngene.chr=5,
    win.size=25,
    KS.cut=0.1,
    sam.name="test",
    distance="euclidean",
    norm.cell.names="",
    output.seg="FLASE",
    plot.genes="TRUE",
    genome="hg20",
    n.cores=1)
```

```r
pred.test <- data.frame(copykat.test$prediction)
CNA.test <- data.frame(copykat.test$CNAmat)
```
