```python
import os
import pandas as pd
import numpy as np

import scanpy as sc
```

```python
counts = sc.read_h5ad('../data/single_cell/original/count_matrix/counts_with_doublet_score.h5ad')
counts
```

```python
out_dir = '../data/single_cell/copycat/inputs/'
```

```python
counts.obs
```

```python
samples = sorted(set(counts.obs['sample_id']))
len(samples)
```

```python
samples
```

```python
for s in samples:
    print(s)
    f = counts[counts.obs['sample_id']==s]
    df = pd.DataFrame(data=f.X.toarray(), columns=f.var.index.to_list(), index=f.obs.index.to_list())
    df = df.transpose()
    df.index.name = 'Gene'
    df.to_csv(os.path.join(out_dir, f'{s}.txt'), sep='\t')
```
