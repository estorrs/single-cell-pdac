```python
htan_sample_map = {
    '1555-tumor': 'htan_01',
    'H_ZY-1105-07': 'htan_02',
    'H_ZY-1160-06': 'htan_03',
    'H_ZY-1174-06': 'htan_04',
    'TWCE-HT055P-S1PAG': 'htan_05',
    'TWCE-HT081P1-XB1': 'htan_06'
}
adata.obs['sample_id'] = [htan_sample_map.get(x, x) for x in adata.obs['sample_id']]
```

```python
# rename in index
import re
def rename_cell_id(cid):
    if 'chan_seng_yue' in cid:
        return cid.replace('_filtered_gene_bc_matrices_h5.h5', '')
    if 'htan' in cid:
        sample = re.sub(r'^htan_surgical_(.*)_[AGCT]+.*$', r'\1', cid)
        converted = htan_sample_map[sample]
        barcode = re.sub(r'^htan_surgical_.*_([AGCT]+).*$', r'\1', cid)
        return f'surgical_{converted}_{barcode}'
    return cid
adata.obs['cell_id'] = [rename_cell_id(x) for x in adata.obs.index]
```
